```python
#Drop to only one artist listed in order to pull API
spotify2023_df['artist_name'] = spotify2023_df['artist(s)_name'].str.replace(',.*', '', regex=True)
spotify2023_df.head()
```

```python
#Drop previous artist(s) column and reorganize dataframe for new artist column
reformat_df = spotify2023_df[['track_name', 'artist_name', 'artist_count', 'released_year',
       'released_month', 'released_day', 'in_spotify_playlists',
       'in_spotify_charts', 'streams', 'in_apple_playlists', 'in_apple_charts',
       'in_deezer_playlists', 'in_deezer_charts', 'in_shazam_charts', 'bpm',
       'key', 'mode', 'danceability', 'valence', 'energy', 'acousticness',
       'instrumentalness', 'liveness', 'speechiness']]
reformat_df
```

```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
import sys
cid = '5b84c854781a447a90292851923f8fb4'
secret = 'd4fee507f6af4cc9b79839a42d57c49a'
client_credentials_manager = SpotifyClientCredentials(client_id=cid, client_secret=secret)
spotify = spotipy.Spotify(client_credentials_manager
=
client_credentials_manager)
```

```python
names = reformat_df['artist_name']
artists = []
genres= []

for name in names: 
    response = spotify.search(q='artist:' + name, type='artist')
    items = response['artists']['items']
    if len(items) > 0:
        artist = items[0]
        singer = artist['name']
        genre = artist['genres']
        artists.append(singer)
        genres.append(genre)
        print(artist['name'], artist['genres'])
    else: print('Not found')
```

```python
results_df = pd.DataFrame({'artist_name': artists, 'genre': genres})
#Change genre from object to string
results_df['genre'] = results_df['genre'].astype(str)
#drop [] around genre lists
results_df['genre'] = results_df['genre'].str.replace('[','').str.replace(']','')
results_df

#29 Artist's were not found
```

```python
#Merge API results with original dataset
merged_df = pd.merge(reformat_df, results_df, on = 'artist_name')
merged_df.drop_duplicates(keep='first', inplace=True)
merged_df
```

```python
merged_df.columns
```

```python
#Reset index and reformat dataframe
merged_df = merged_df[['track_name', 'artist_name', 'genre', 'artist_count', 'released_year',
       'released_month', 'released_day', 'in_spotify_playlists',
       'in_spotify_charts', 'streams', 'in_apple_playlists', 'in_apple_charts',
       'in_deezer_playlists', 'in_deezer_charts', 'in_shazam_charts', 'bpm',
       'key', 'mode', 'danceability', 'valence', 'energy', 'acousticness',
       'instrumentalness', 'liveness', 'speechiness']]
merged_df.reset_index(inplace = True, drop = True)
merged_df
```

```python
#Drop to first genre Spotify lists
merged_df['genre'] = merged_df['genre'].str.replace(',.*', '', regex=True)
merged_df
```
