```python
import pandas as pd
from pathlib import Path
import sqlite3
```

```python
# Store filepath in a variable
spotify2023 = Path("spotify-2023.csv")
```

```python
# Read our data file with the Pandas library

spotify2023_df = pd.read_csv('spotify-2023.csv', encoding='ISO-8859-1')
```

```python
spotify2023_df
```

```python
spotify2023_df.dtypes
```

```python
# Convert all values in the "streams" column to strings
spotify2023_df['streams'] = spotify2023_df['streams'].astype(str)

# Filter rows where the "streams" column contains only numeric values
spotify2023_df = spotify2023_df[spotify2023_df['streams'].str.isnumeric()]

# Convert the "streams" column to int64
spotify2023_df['streams'] = spotify2023_df['streams'].astype('int64')
```

```python
# Convert all values in the "in_deezer_playlists" column to strings
spotify2023_df['in_deezer_playlists'] = spotify2023_df['in_deezer_playlists'].astype(str)

# Filter rows where the "in_deezer_playlists" column contains only numeric values
spotify2023_df = spotify2023_df[spotify2023_df['in_deezer_playlists'].str.isnumeric()]

# Convert the "in_deezer_playlists" column to int64
spotify2023_df['in_deezer_playlists'] = spotify2023_df['in_deezer_playlists'].astype('int64')
```

```python
# Convert all values in the "in_shazam_charts" column to strings
spotify2023_df['in_shazam_charts'] = spotify2023_df['in_shazam_charts'].astype(str)

# Filter rows where the "in_deezer_playlists" column contains only numeric values
spotify2023_df = spotify2023_df[spotify2023_df['in_shazam_charts'].str.isnumeric()]

# Convert the "in_deezer_playlists" column to int64
spotify2023_df['in_shazam_charts'] = spotify2023_df['in_shazam_charts'].astype('int64')
```

```python
spotify2023_df.rename(columns = {'danceability_%':'danceability', 
                                  'valence_%':'valence','energy_%': 'energy', 'acousticness_%':'acousticness',
                                 'instrumentalness_%':'instrumentalness','liveness_%':'liveness','speechiness_%':'speechiness'
                                }, inplace = True)
```
